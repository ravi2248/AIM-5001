```python
#importing required libraries
import pandas as pd
```

```python
#Here we are reading the csv file which is in the github repository
data = pd.read_csv('https://raw.githubusercontent.com/ravi2248/AIM-5001/main/Goods%20%2B%20Services%20(1).csv', names=["Kind", "Type", "2019", "2020", "2021", "Category"], header=None)
```

```python
#we can see the data
data
```

```python
#we are droping the rows which has all null values
data = data.dropna(axis=0, how="all")
```

```python
data
```

```python
#Here we are making var_name has year and value_name as volume in melt function
data = pd.melt(data, id_vars=["Kind", "Type", "Category"], var_name="Year", value_name="Volume")
```

```python
data
```

```python
#Here we are dropping the volume rows which has null values
data = data.dropna(subset=["Volume"])
```

```python
#Here we are dropping the rows having kind, type, and category columns are null
data = data.dropna(subset=["Kind", "Type", "Category"], how = 'all')
```

```python
data
```

```python
#we are resetting the indexes
data.reset_index(drop=True, inplace=True)
```

```python
#Here we are dropping the category column which has null values
data = data.drop("Category", axis=1)
```
